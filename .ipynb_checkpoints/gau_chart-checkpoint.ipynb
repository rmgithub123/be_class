```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
new_c = pd.read_csv("/home/madhuri/Downloads/experiment.csv")
```

```python
new_c
```

```python
plt.subplot(3,2,1)

sns.lineplot(x='Months',y='Transport',data=new_c)
plt.ylabel('Transport')
plt.subplot(3,2,2)
sns.lineplot(x='Months',y='tra_per',data=new_c)
plt.ylabel('Transport Percentage')
plt.subplot(3,2,3)
sns.lineplot(x='Months',y='ETA',data= new_c)
plt.ylabel('ETA')
plt.subplot(3,2,4)
sns.lineplot(x='Months',y='eta_per',data= new_c)
plt.ylabel('ETA%')

plt.subplot(3,2,5)
sns.lineplot(x='Months',y='POD',data= new_c)
plt.ylabel('POD')
plt.subplot(3,2,6)
sns.lineplot(x='Months',y='pod_per',data= new_c)
plt.ylabel('POD%')
plt.show()
```
