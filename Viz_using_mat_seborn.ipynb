## Seaborn 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
type(tips)
```

```python
tips
```

## Univarite - one variable data visualization

```python
#strip  plot

sns.stripplot(y='tip',data=tips, jitter = True)
plt.ylabel('tip $')
plt.show()
```

```python
#Grouping with Stripplot

sns.stripplot(x='day',y='tip',data=tips,hue ='day',jitter =False)
plt.ylabel('tips$')
plt.xlabel('Day of the week')
plt.show()
```

## Swarmplot

```python
sns.swarmplot(x='day',y='tip', data=tips, hue = 'day')
plt.ylabel('tip $')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip', data=tips, hue = 'sex')
plt.ylabel('tip $')
plt.show()
```

```python
sns.swarmplot(x='tip',y='day', data=tips,hue = 'sex',orient='h')
plt.xlabel('tip $')
plt.show()
```

## Box and Voilin Plot

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip $')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip $')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip $')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip $')

plt.show()
```

```python
# combining plots

sns.violinplot(x='day',y='tip', data=tips, inner = None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4 ,hue='sex')
plt.ylabel('tip $')
plt.show()
```

## Bivariate -> "Two variable " Data Visualization

```python
#Joint PLot 

sns.jointplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
#Density plot

sns.jointplot(x='total_bill', y = 'tip', data=tips, kind='hex')
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
plt.show()
```
