```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
# merged_df.to_csv('merged_data.csv', index=False)  
```

```python
movies_df.head()
```

```python
merged_df = pd.merge(movies_df,ratings_df, on='movieId', how='inner')
```

```python
# Define a user for whom you want to make recommendations
user_id = 1  # You can replace this with any user ID from your dataset
```

```python
# Filter the dataset to get the movies the user has already rated
#user_ratings = merged_df[merged_df['userId'] == user_id]
user_ratings = merged_df
user_ratings.shape
user_ratings.head()
```

```python
# Create a user-movie matrix with ratings as values
user_movie_matrix = user_ratings.pivot(index='userId', 
                                       columns='movieId', 
                                       values='rating').fillna(0)
user_movie_matrix.head()
```

```python
user_movie_matrix.shape
```

```python
user_sim = 1 - pairwise_distances(user_movie_matrix.values, metric = 'cosine')
```
