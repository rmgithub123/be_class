```python
movies_df = pd.read_csv('/home/madhuri/Downloads/ml-latest-small/movies.csv')
ratings_df = pd.read_csv('/home/madhuri/Downloads/ml-latest-small/ratings.csv')
```

```python
merged_df = pd.merge(movies_df,ratings_df, on='movieId', how='inner')
```

# Define a user for whom you want to make recommendations


```python
user_id = 1  # You can replace this with any user ID from your dataset
```

```python
# Filter the dataset to get the movies the user has already rated
user_ratings = merged_df[merged_df['userId'] == user_id]
```

```python
# Find movies that the user has not yet rated
unrated_movies = merged_df[~merged_df['movieId'].isin(user_ratings['movieId'])]
```

```python
unrated_movies.head()
```

```python
# Group the unrated movies by 'movieId' and calculate the mean rating for each movie
movie_ratings = unrated_movies.groupby(['movieId','title'])['rating'].mean().reset_index()
```

```python
movie_ratings
```

```python
# Sort the movies by mean rating in descending order to recommend the highest-rated ones
recommended_movies = movie_ratings.sort_values(by='rating', ascending=False)
```

```python
# Display the top N recommended movies (e.g., top 10)
top_n_recommendations = recommended_movies.head(10)
```

```python
# Print the top recommendations
print("Top 10 Movie Recommendations for User", user_id)
print(top_n_recommendations)
```
